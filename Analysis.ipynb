```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

from matplotlib import rcParams
rcParams.update({'font.size': 22})

from pickle import load
with open('radSer.pkl','rb') as f:
    results = load(f)
```

```python
times = [results[i][:,0] for i in range(99)]
servo1 = [results[i][:,1] for i in range(99)] #increases with time
servo2 = [results[i][:,2] for i in range(99)] #decreases with time
servo3 = [results[i][:,3] for i in range(99)] #increases with time
dists = [results[i][:,5] for i in range(99)]
```

# Plots

### Distance as a function of Servo 3

```python
plt.figure(figsize=(16,9), dpi=100)
[plt.plot(servo3[i], dists[i], '-o') for i in range(99)];
plt.xlabel('Servo 3 Position')
plt.ylabel('Distance from Target (meters)')
plt.xlim((14975,17300))
plt.grid('on')
plt.tight_layout()
plt.savefig('DistanceVsServo3_Many.png')
```

### Distance as a function of Servo 2

```python
plt.figure(figsize=(16,9), dpi=100)
[plt.plot(servo2[i], dists[i], '-o') for i in range(99)];
plt.xlabel('Servo 2 Position')
plt.ylabel('Distance from Target (meters)')
plt.xlim((14050,16450))
plt.grid('on')
plt.tight_layout()
plt.savefig('DistanceVsServo2_Many.png')
```

#### Averaging Data

```python
data = [zip(servo3[i],dists[i]) for i in range(99)]
d = []
for i in data:
    d+=i
data = sorted(d)
```

```python
dataToPlot = np.array([np.mean(data[128*i:128*i+129],axis=0) for i in range(22)])
stdToPlot = np.array([np.std(data[128*i:128*i+129],axis=0) for i in range(22)])
```

### Average distance a funciton of Servo 3

```python
plt.figure(figsize=(16,9), dpi=100)
plt.plot(dataToPlot[:,0],dataToPlot[:,1],'-o')
plt.xlabel('Servo 3 Position')
plt.ylabel('Distance from Target (meters)')
plt.xlim((14950,17100))
plt.grid('on')
plt.tight_layout()
plt.savefig('DistanceVsServo3_Average.png')
```

### Same as above but with error bars due to averaging

```python
plt.figure(figsize=(16,9), dpi=100)
plt.errorbar(dataToPlot[:,0], dataToPlot[:,1], xerr=stdToPlot[:,0], yerr=stdToPlot[:,1],fmt='-o')
plt.xlabel('Servo 3 Position')
plt.ylabel('Distance from Target (meters)')
plt.xlim((14850,17300))
plt.grid('on')
plt.tight_layout()
plt.savefig('DistanceVsServo3_AverageErrorBars.png')
```
